**# =============================================== Q.34 =============================================== #**

Q34. For each city, compute three metrics in a single groupby using .agg() with a dictionary:
(a) Number of startups
(b) Total funding raised (sum of amount)
(c) Average funding per startup
Print the top 10 cities by number of startups.

```python
# Compute city-wise startup statistics
# .agg() - performs multiple operations at once in single groupby() operation ...(like sum, mean, count, etc.)

city_stats = df.groupby("city").agg({
    "startup_name": "count",
    "amount": ["sum", "mean"]
})

city_stats.columns = [    # Rename the columns
    "startup_count",
    "total_funding",
    "average_funding"
]

# Sort by startup count in descending order
city_stats = city_stats.sort_values(
    by="startup_count",
    ascending=False
)

print(city_stats.head(10))     # Display top 10 cities
```

Q39. Investor hub analysis. A foreign investor wants to know which Indian city has the highest TOTAL
funding raised, considering only cities with at least 20 startups (so the number is meaningful). Print city
name, count, and total funding.

```python
# ----- Investor Hub List ----- #

# Calculate startup count and total funding for each city
city_stats = df.groupby("city").agg({
    "startup_name": "count",
    "amount": "sum"
})

city_stats.columns = [       # Rename the columns
    "startup_count",
    "total_funding"
]
```

```python
# Keep cities with at least 20 startups
city_stats = city_stats[  city_stats["startup_count"] >= 20  ]

# Sort by total funding in descending order
city_stats = city_stats.sort_values(
    by="total_funding",
    ascending=False
)

# Display the city with the highest total funding
print(city_stats.head(1))
```

Q40. Market saturation. For each combination of (city, industry), count the number of startups. Print the
top 10 most-saturated combinations. This tells a new founder where competition is most intense – and
by elimination, where the gaps might be.

```python
# -------- MARKET SATURATION -------- #

# cond-1 : df.groupby(["city", "industry"])    ...groupby 2 columns
# cond-2 : counr startups ... startup_count
# cond-3 : sort and print top 10

# Count startups for each city and industry combination
market_saturation = df.groupby(
    ["city", "industry"]
).agg({
    "startup_name": "count"
})

market_saturation.columns = ["startup_count"]        # Rename the column
```

```python
# Sort by startup count in descending order
market_saturation = market_saturation.sort_values(
    by="startup_count",
    ascending=False
)

# Display the top 10 combinations
print(market_saturation.head(10))
```
